```python
import matplotlib.pyplot as plt
from matplotlib import gridspec
from autograd import numpy as np 
import copy
```

```python
def plot_compare(seq1,seq2,seq3,seq4):   
    # initialize figure
    fig = plt.figure(figsize = (15,10))
 
    # create subplot with 3 panels, plot input function in center plot
    gs = gridspec.GridSpec(1,1) 
    ax1 = plt.subplot(gs[0]);
 
    # plot
    ax1.plot(np.arange(np.size(seq1)),seq1.flatten(),c = 'k',linewidth = 2.5,linestyle ='--',label='raw')
    ax1.plot(np.arange(np.size(seq2)),seq2.flatten(),c = 'b',linewidth = 2.5,linestyle ='-',label='MAF')
    ax1.plot(np.arange(np.size(seq3)),seq3.flatten(),c = 'r',linewidth = 2.5,linestyle ='-',label='IIR')
    ax1.plot(np.arange(np.size(seq4)),seq4.flatten(),c = 'g',linewidth = 2.5,linestyle ='-',label='FIR')
    # label axes and title
    ax1.set_title('filters comparison')
    ax1.set_xlabel('step')
    
    ax1.legend()
    
    # set viewing limits
    s1min = np.min(copy.deepcopy(seq1))
    s1max = np.max(copy.deepcopy(seq1))
    s1gap = (s1max - s1min)*0.1
    s1min -= s1gap
    s1max += s1gap
    ax1.set_ylim([s1min,s1max])
    
    
 
    plt.show()
```

```python
data = np.loadtxt('data.csv',delimiter = ',')
x=data[:,0]
raw=data[:,1]
maf=data[:,2]
iir=data[:,3]
fir=data[:,4]
```

```python
plot_compare(raw,maf,iir,fir)
```
